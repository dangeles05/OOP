## Calcular maximo

Escribir una funcion `maximum()` que calcule el maximo de una lista de valores

```python
def maximum(values):
    answer = None
    for value in values:
        if answer == None or answer < value:
            answer = value
    return answer
```

```python
def random_inputs(length):
    return [random.randint(-1000, 10000) for _ in range(length)]
```

## Testing funcion `maximum()`

```python
values = [-2, 4,6,7,8]
maximo = maximum(values)
print(maximo) #8
```

```python
values = random_inputs(10)
print(values)
start = time.time()
maximo = maximum(values)
end = time.time()

print(end - start)
print(maximo) #8
```

```python
runtime = []
for length in range(1, 501):
    values = random_inputs(length)
    start = time.time()
    maximo = maximum(values)
    end = time.time()
    runtime.append(end - start)
print('Ejecucion terminada')
plot_times(runtime)
```

# Caracterizacion de la complejidad de un algoritmo

```python
def maximum(values):
    answer = None                                       #c1, 1 time
    for value in values:                                #c2, n times
        if answer == None or answer < value:            #c3, n times
            answer = value                              #c4, n times
    return answer                                       #c5, 1 time
```

La complejidad de `maximum` es:

$
\begin{align}
\text{runtime} & = c_1 + c_2 + c_3 + c_4 + c_5\\
               & = 1   + N   + N   + N   + 1\\
               & = 3N  + 2
\end{align}
$

Crear una seccion `sum_all()` que sume todos los valores de una lista

```python
def sum_all(values):
    total = 0
    for value in values:
        total += value
    return total
```

```python
def sum_all(values):
    total = 0                                           # c1, 1 time
    for value in values:                                # c2, N times
        total += value                                  # c3, N times
    return total                                        # c4, 1 time
```

Crear una funcion `sum_zeros()` que dadas dos listas de valores de la misma longitud cuente todos los pares que sumen cero

```python
def sum_zeros(list1, list2):
    count = 0
    for i in range(len(list1)):
        if list1[i] + list2[i] == 0:
            count += 1
    return count
```

```python
list1 = [2, 5, 3, -4]
list2 = [-2, 7, 6, 4]
value = sum_zeros(list1, list2)
print(value)  # 3
```

Crear una funcion `min_max()` que calcule el mínimo y el máximo de una lista de valores

```python
def min_max(values):
        minimum = values[0]
        maximum = values[0]

        for value in values:
            if value < minimum:
                minimum = value
            if value > maximum:
                maximum = value
        return minimum, maximum
```

```python
values = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
resultado = min_max(values)
print(f"Mínimo: {resultado[0]}, Máximo: {resultado[1]}")
```

# Complejidad de `min_max()`

```python
def min_max(values):
        minimum = values[0]                             # c1, 1 time
        maximum = values[0]                             # c2, 1 time
        for value in values:                            # c3, N times
            if value < minimum:                         # c4, N times
                minimum = value                         # c5, N times
            if value > maximum:                         # c6, N times
                maximum = value                         # c7, N times
        return minimum, maximum                         # c8, 1 time
```

La complejidad de `min_max` es:
$
\begin{align}
\text{runtime} & = c_1 + c_2 + c_3 + c_4 + c_5 + c_3 + c_4 + c_5\\
               & = 1   + 1   + N   + N   + N   + N   + N   + 1   \\
               & = 5N  + 3
\end{align}
$

# Crecimiento no tiempo de ejecución  

Solo interesa ver cuanto crece el tiempo de ejecución a medida que crecen los datos, no el tiempo de ejecución exacto

```python
N = np.linspace(1, 100000)
N_3 = N**3
N_2 = 5*N**2
N_17 = 17*N
plot_times(N_3, N_2, N_17, N_3 + N_2 + N_17)
```

$
\begin{align}  
                & \mathcal{O}(2N^{2} + N + 3) = \mathcal{O}(N^{2})\\
                & \mathcal{O}(3N + 2) = \mathcal{O}(N)\\
                & \mathcal{O}(N\log{N}+ 1) = \mathcal{O}(N\log{N})\\
                & \mathcal{O}(3N^{2} + N^{2})\\
                & \mathcal{O}(N^{2})\\
\end{align}
$
