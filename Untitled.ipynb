```python
import matplotlib.pyplot as plt
import random
import time
```

```python
def plot_times(times):
    fig, ax = plt.subplots(figsize=(12,8))

    ax.plot(times, linewidth=2, color="r", linestyle="solid", 
            marker=">", markersize=12, markerfacecolor="b", 
            markeredgecolor="gray")

    ax.grid(True)  # Se activa la cuadrícula sobre el gráfico
    plt.show()  # Se muestra el gráfico
```

## Calcular maximo

Escribir una funcion `maximum()` que calcule el maximo de una lista de valores

```python
def maximum(values):
    answer = None
    for value in values:
        if answer == None or answer < value:
            answer = value
    return answer
```

```python
def random_inputs(length):
    return [random.randint(-1000, 10000) for _ in range(length)]
```

## Testing funcion `maximum()`

```python
values = [-2, 4,6,7,8]
maximo = maximum(values)
print(maximo) #8
```

```python
values = random_inputs(10)
print(values)
start = time.time()
maximo = maximum(values)
end = time.time()

print(end - start)
print(maximo) #8
```

```python
runtime = []
for length in range(1, 501):
    values = random_inputs(length)
    start = time.time()
    maximo = maximum(values)
    end = time.time()
    runtime.append(end - start)
print('Ejecucion terminada')
plot_times(runtime)
```
