```python
class Time:
    def __init__(self, hour, minu, sec):
        self.hour = hour
        self.minu = minu
        self.sec = sec

    def total_seconds(self):
        return self.hour * 3600 + self.minu * 60 + self.sec

    def __str__(self):
        return (f'{self.hour}:{self.minu}:{self.sec} hrs')
```

```python
ti = Time(16,33,11)
print(ti)
ti.total_seconds()
```

```python
class FrecqTable:
    def __init__(self):
        self.count = {}

    def add(self, a):
        if a in self.count:
            self.count[a] += 1
        else:
            self.count[a] = 1
```

```python
freq = FrecqTable(1)
print(freq.count)
```
